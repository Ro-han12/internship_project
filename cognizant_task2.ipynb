```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):
    
  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

```python
data1 = convert_to_datetime(data1, 'timestamp')
data1.info()
```

```python
data2 = convert_to_datetime(data2, 'timestamp')
data2.info()
```

```python
data3 = convert_to_datetime(data3, 'timestamp')
data3.info()
```

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
```

```python
data1 = convert_timestamp_to_hourly(data1, 'timestamp')
data1.head()
```

```python
data2 = convert_timestamp_to_hourly(data2, 'timestamp')
data2.head()
```

```python
data3 = convert_timestamp_to_hourly(data3, 'timestamp')
data3.head()
```
